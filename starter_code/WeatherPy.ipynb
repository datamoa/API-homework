```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{base_url}appid={api_key}&units={units}&q="
print(query_url)
```

```python
# #Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude
temp = []
humidity = []
cloudiness = []
wind_speed = []
lat = []
lon = []
date = []
country = []
dr = []
city_set=1
counter = 0
#Skip cities not found in OpenWeatherMap and make the code run.
#Retrive and add all related data
for city in  cities:
    counter = counter +1
    try: 
        response = requests.get(query_url + city).json()
        dr.append(response["name"])
        temp.append(response['main']['temp'])
        humidity.append(response["main"]['humidity'])
        cloudiness.append(response["clouds"]["all"])
        wind_speed.append(response['wind']["speed"])  
        lat.append(response["coord"] ['lat'])
        lon.append(response["coord"] ['lon'])
        date.append(response["dt"])
        country.append(response["sys"]["country"])
        print("Processing Record " + str(counter) + " of set " + str(city_set) +  " | " + city)
     
    except:
        print("City not found. Skipping: " + city)
```

```python
print("{:<15}".format("City"), len(humidity))
print("{:<15}".format("Temperature"), len(temp))
print("{:<15}".format("Cloudiness"), len(cloudiness))
print("{:<15}".format("Wind Speed"), len(wind_speed))
print("{:<15}".format("Latitude"), len(lat))
print("{:<15}".format("Longtitude"), len(lon))
print("{:<15}".format("Date"), len(date))
print("{:<15}".format("Country"), len(country))
```

```python
summary = {
    "City": dr,
    "Country": country,
    "Date": date,
    "Temperature": temp,
    "Latitude": lat,
    "Longtude": lon,
    "Humidity": humidity,
    "Wind speed": wind_speed,
    "Cloudiness": cloudiness
    }
summary = pd.DataFrame(summary)
summary.head()
```
